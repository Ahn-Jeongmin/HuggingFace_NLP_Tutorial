```python
#(모델 로딩 방법 1) 기본 아키텍처 가져오기

from transformers import BertConfig, BertModel

# Building the config
config = BertConfig()

# Building the model from the config
model = BertModel(config)

#구성을 라이브러리에서 받아오고, 그 구성을 모델에 매개변수로 전달
#이와 같은 방법은 모델의 초기화를 랜덤으로 진행 ! -> 그러나 아키텍처만 제공할 뿐 따로 훈련을 진행해야
```

```python
print(config) #모델에 전달할 구성 확인

#hidden_size 는 hidden state 벡터의 크기를 정의
# num_hidden_layers 는 트랜스포머 모델이 가지는 레이어의 개수를 지정
```

```python
#(모델 로딩 방법 2) 이미 학습된 pretrained 모델 로드하기

from transformers import BertModel

model = BertModel.from_pretrained("bert-base-cased")
```

```python
print(model)
```

```python
#모델 저장 -> save_pretrained() 메서드 활용
#config 정보와 모델 파일 -> 총 두 가지 파일을 컴퓨터의 디렉토리에 저장
# config.json 파일과 pytorch_model.bin

model.save_pretrained("directory_on_my_computer")
```

```python
#  *********** 전체 과정 리뷰 ***********

from transformers import BertTokenizer

tokenizer = BertTokenizer.from_pretrained("bert-base-cased")
sequences = ["Hello!", "Cool.", "Nice!"]
inputs = tokenizer(sequences)

print(inputs)
```

```python
encoded_sequences = inputs['input_ids']
print(encoded_sequences)
```

```python
import torch

# transformer는 텐서 입력만 받기 때문에 input 데이터를 tensor로의 변환을 해 주는 작업이 필수적이다
model_inputs = torch.tensor(encoded_sequences)
print(model_inputs)
```

```python
#텐서로 변환된 input 데이터를 model의 매개변수로 전달해주기만 하면 된다

output = model(model_inputs)
print(output)
```
